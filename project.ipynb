```python
import gymnasium as gym
from stable_baselines3 import PPO
import os
```

```python
# Tworzymy środowisko BipedalWalker
# render_mode="human" pozwoli nam widzieć okno z grą
env = gym.make("BipedalWalker-v3", render_mode="human")

# Resetujemy środowisko
observation, info = env.reset()

print("Rozpoczynam spacer testowy...")

for _ in range(200):  # Robimy 200 kroków
    # Losowa akcja (ruszaj nogami byle jak)
    action = env.action_space.sample()
    
    # Wykonaj krok
    observation, reward, terminated, truncated, info = env.step(action)
    
    if terminated or truncated:
        observation, info = env.reset()

env.close()
print("Koniec testu.")
```
